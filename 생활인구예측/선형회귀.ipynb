```python
# 데이터 합침
df_total = pd.concat([df17,df18,df19,df20,df21])
```

```python
df_total = df_total[['기준일ID','시간대구분','총생활인구수']]
df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'], format='%Y%m%d')
df_total
```

```python
df_total['date'] = pd.to_datetime(df_total['기준일ID'])
df_total['year'] = df_total['date'].dt.year
df_total['month'] = df_total['date'].dt.month
df_total['day'] = df_total['date'].dt.day
```

```python
df_total.rename(columns={'시간대구분':'hour'},inplace=True)
df_total = df_total[['year','month','day','hour','총생활인구수']]
df_total
```

```python
# index 재배열
df_total.reset_index(inplace=True, drop=True)
```

```python
df_total.to_csv('df_total_final.csv', index=False)
```

```python
# 일주일 단위로 그래프 확인 
plt.figure(figsize=(20,4))

plt.plot(df19['총생활인구수'][:24*7], data = df19)
plt.plot(df19['총생활인구수'][24*7:24*7*2], data = df19, color='red')
plt.plot(df19['총생활인구수'][24*7*2:24*7*3], data = df19, color='green')
plt.plot(df19['총생활인구수'][24*7*3:24*7*4], data = df19, color='gray')

# 일주일 단위로 확인했을 때 그래프의 모양이 비슷한 것을 확인 
# 따라서 요일별로 그래프의 모양이 있다고 생각하여 찾으려는 2022년의 1월 1일이 '토'요일인 것을 감안하여
# 구하려는 1416개의 데이터가 아닌 1416 + 96 을 하여 '토'요일 데이터에 시작점을 맞춘 1512개의 데이터를 shift하기로 결정
```
